Datasheet 4.04B: Area of destination or origin within the UK by citizenship

```python
from databaker.framework import *
import pandas as pd
```

```python
df1 = pd.read_excel(xls, 'Datasheet 4.04B', header = 7)
```

```python
df1
```

```python
observations = df1.iloc[:, 4:]
observations.head()
```

```python
new_table = pd.melt(observations,
                       ["Row Label"],
                       var_name="Column label",
                       value_name="Value")
```

```python
new_table.head(50)
```

```python
new_table.shape
```

```python
new_table.dropna(inplace=True)
```

```python
new_table = new_table[new_table['Value'] != '.']
```

```python
new_table = new_table[new_table['Value'] != '0']
```

```python
new_table.head(50)
```

```python
new_table.count()
```

```python
new_table['Country of Residence'] = new_table['Column label'].replace(to_replace= r"CIT (.*)(EST|CI )$", value = "\\1", regex=True)
new_table['measure'] = new_table['Column label'].replace(to_replace= r".* (EST|CI )$", value = "\\1", regex=True)
new_table['Country of Residence'] = new_table['Country of Residence'].str.lower()
new_table['Country of Residence'] = new_table['Country of Residence'].str.strip()
new_table['measure'] = new_table['measure'].str.lower()
new_table['measure'] = new_table['measure'].str.strip()
new_table['Row Label'] = new_table['Row Label'].str.lower()
temp_table = new_table['Row Label'].str.split(',', -1, expand=True)
new_table['Flow'] = temp_table[0]
new_table['Area of Destination or Origin'] = temp_table[1]
new_table['Flow'] = new_table['Flow'].str.strip()
new_table['Area of Destination or Origin'] = new_table['Area of Destination or Origin'].str.strip()
new_table['Geography'] = 'K02000001'
new_table['Unit'] = 'People (thousands)'
new_table['Measure Type'] = 'Count'
new_table['Year'] = '2016'
temp_table = new_table[:]
temp_table['Keys'] = temp_table['Row Label'] + '/' + temp_table['Column label']
temp_table['Keys'] = temp_table['Keys'].str.strip()
temp_table['Keys2'] = temp_table['Keys'].str.rstrip('EST')
temp_table['Keys2'] = temp_table['Keys2'].str.rstrip('CI')
temp_table['Keys2'] = temp_table['Keys2'].str.strip()
Final_table = pd.merge(temp_table, temp_table[['Keys2','Value','measure']], how = 'left', left_on = 'Keys2', right_on = 'Keys2')
Final_table = Final_table[Final_table['measure_x'] != Final_table['measure_y']]
Final_table = Final_table.drop_duplicates('Keys2', keep = 'first')
Final_table.columns = ['Value' if x=='Value_x' else x for x in Final_table.columns]
Final_table.columns = ['CI' if x=='Value_y' else x for x in Final_table.columns]
Final_table['Value'] = Final_table['Value'].astype(float)
Final_table['Year'] = Final_table['Year'].astype(int)
Final_table['CI'] = Final_table['CI'].astype(float)
Final_table = Final_table[Final_table['Country of Residence'] != 'all']
Final_table = Final_table[['Geography','Year','Country of Residence','Flow','Area of Destination or Origin','Measure Type','Value','CI','Unit']]
```
